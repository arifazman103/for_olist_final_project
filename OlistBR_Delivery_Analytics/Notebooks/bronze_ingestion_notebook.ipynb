```python
orders_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/raw_csv/olist_orders_dataset.csv")

display(orders_df)

orders_df.write.mode("overwrite").format("delta").saveAsTable("bronze_orders")
```

```python
products_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/raw_csv/olist_products_dataset.csv")

products_df = products_df.withColumnRenamed("product_name_lenght", "product_name_length")
products_df = products_df.withColumnRenamed("product_description_lenght", "product_description_length")

display(products_df)

products_df.write.mode("overwrite").option("overwriteSchema", "true").format("delta").saveAsTable("bronze_products")
```
